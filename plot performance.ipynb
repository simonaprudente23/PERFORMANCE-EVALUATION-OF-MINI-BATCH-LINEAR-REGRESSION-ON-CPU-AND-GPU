```python
import numpy as nm
import matplotlib.pyplot as plt
times_gpu = nm.array([284.508 , 68.188 , 288.542])
times_cpu = nm.array([276 , 60 , 383])
dataset = nm.array([60000000,  13500000, 202500000])
```

```python
from scipy.interpolate import interp1d
plt.figure(figsize = [10, 8])
#plt.scatter(dataset, times_cpu)
#plt.scatter(dataset, times_gpu)
f = interp1d(dataset, times_cpu, fill_value="extrapolate")
f1 = interp1d(dataset, times_gpu, fill_value="extrapolate")

xnew = nm.linspace(0, 202500000, num=100, endpoint=True)
plt.plot(dataset, times_cpu, 'o', xnew, f(xnew), dataset, times_gpu, 'x', xnew, f1(xnew))
#plt.plot(xnew, f(xnew))
plt.ylabel('seconds')
plt.xlabel('number of datapoints')
plt.yscale('log')
plt.title("Difference of performances between CPU and GPU", fontsize = '14')
plt.legend(['performance_cpu_scatter','performance_cpu_interpolation','performance_gpu_scatter', "performance_gpu_interpolation"])
```
